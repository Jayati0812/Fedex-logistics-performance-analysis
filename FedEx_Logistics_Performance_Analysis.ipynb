## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
data.select_dtypes(include=['int64','float64'])
````

````python
data.select_dtypes(include=['object'])
````

````python
#converting the date column from object data type to date
data['Delivered to Client Date']=pd.to_datetime(data['Delivered to Client Date'])
data['Delivery Recorded Date']=pd.to_datetime(data['Delivery Recorded Date'])
data['PO Sent to Vendor Date']=data['PO Sent to Vendor Date'].replace("Date Not Captured",np.nan) #replaces date not captured with Nan
data['PO Sent to Vendor Date']=pd.to_datetime(data['PO Sent to Vendor Date'],errors='coerce') #ignores Nan while converting to date
data['Scheduled Delivery Date']=pd.to_datetime(data['Scheduled Delivery Date'])
data['PQ First Sent to Client Date']=pd.to_datetime(data['PQ First Sent to Client Date'],errors='coerce')  #ignores Nan while converting to date
````

````python
#cleaning the column Weight (Kilograms)
# print(data['Weight (Kilograms)'].dtype)
data['Weight (Kilograms)']=data['Weight (Kilograms)'].replace("Weight Captured Separately",np.nan)  #replaces weight captured separately with Nan
data.loc[data['Weight (Kilograms)'].str.contains('DN|See', na=False), 'Weight (Kilograms)'] = np.nan #replaces DN or See with Nan
data['Weight (Kilograms)']=data['Weight (Kilograms)'].astype('float64')   #converts to float type
data['Weight (Kilograms)'].fillna(data['Weight (Kilograms)'].median(), inplace=True)  #replaces with Nan with median
print(data.loc[:,'Weight (Kilograms)'])
````

````python
#cleaning the column Freight Cost (USD)
data['Freight Cost (USD)']=data['Freight Cost (USD)'].replace(["Freight Included in Commodity Cost","Invoiced Separately"],np.nan) #replaces Freight Included in Commodity Cost and Invoiced Separately with Nan
data.loc[data['Freight Cost (USD)'].str.contains('DN|See', na=False), 'Freight Cost (USD)'] = np.nan #replaces DN or See with Nan
data['Freight Cost (USD)']=data['Freight Cost (USD)'].astype('float64') #converts to float type
data['Freight Cost (USD)'].fillna(data['Freight Cost (USD)'].median(), inplace=True)  #replaces with Nan with median
print(data.loc[:,'Freight Cost (USD)'])
````

````python
#replacing the null values in Line Item Insurance (USD)
data['Line Item Insurance (USD)'].isnull().sum()
data['Line Item Insurance (USD)'].fillna(0,inplace=True)  #replaces with Nan with 0
data['Line Item Insurance (USD)'].isnull().sum()
print(data['Line Item Insurance (USD)'])
````

````python
#replacing the null values in Shipment Mode
data['Shipment Mode'].isnull().sum()
data['Shipment Mode'].fillna(data['Shipment Mode'].mode()[0],inplace=True)    #replaces with Nan with mode of shipment mode
data['Shipment Mode'].isnull().sum()
print(data['Shipment Mode'])
````

````python
#replacing the null values in Dosage
data['Dosage'].isnull().sum()
data['Dosage'].fillna(data['Dosage'].mode()[0],inplace=True)     #replaces with Nan with mode of shipment mode
data['Dosage'].isnull().sum()
print(data['Dosage'])
````

````python
#creating new column for analysis
data['Delay'] = (data['Delivered to Client Date'] - data['Scheduled Delivery Date']).dt.days
data['Delivery Status'] = data['Delay'].apply(lambda x: 'Late' if x > 0 else 'On Time')
data['Month'] = data['Delivered to Client Date'].dt.to_period('M')
````

To make the dataset analysis-ready, I performed several data cleaning, preprocessing, and feature engineering operations.

#### 1. Date Standardization

#### 2. Handling Missing and Invalid Values

The dataset contained several inconsistencies in logistics-related numerical columns.

**Weight (Kilograms):**

* Removed non-numeric entries containing references like "DN" or "See DN-XXXX".
* Converted the column to numeric format.
* Imputed missing values using the median weight to minimize the impact of outliers.

**Freight Cost (USD):**

* Replaced entries such as "Freight Included in Commodity Cost" and "Invoiced Separately" with null values.
* Removed non-numeric references containing "DN" or "See".
* Converted the column to numeric format.
* Filled missing values using the median freight cost.

**Line Item Insurance (USD):**

* Missing insurance values were replaced with 0, assuming no insurance cost was recorded.

#### 3. Handling Missing Categorical Values

For categorical features:

This preserved the maximum number of records for analysis.

#### 4. Feature Engineering

To support logistics performance analysis, I created new business-focused features:

**Delay**

* Calculated as the difference between Delivered to Client Date and Scheduled Delivery Date.
* Positive values indicate late deliveries.

**Delivery Status**

* Categorized shipments as:

  * Late
  * On Time

**Month**

* Extracted month and year from delivery dates for trend analysis and time-series visualizations.
